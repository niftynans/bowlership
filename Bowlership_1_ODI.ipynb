```python
import powerlaw
import pandas as pd
import numpy as np
import yaml
import math
import os
import matplotlib.pyplot as plt
```

```python
match_list=[]
directory = "/home/pnans/Documents/Projects/Bowlership/Data/odis/"
for filename in os.listdir(directory):
    if filename.endswith(".yaml"): 
        print(os.path.join(directory, filename))
        f = (os.path.join(directory, filename))
        with open(f, 'r') as stream:
            try:
                match_dict = yaml.safe_load(stream)
                match_list.append(match_dict)
                print("loaded")
            except yaml.YAMLError as exc:
                print(exc)
```

```python
over_dict = []
for i in range(len(match_list)):
    d = match_list[i]["innings"][0]['1st innings']['deliveries']
    
    entry_dict = {}
#entry_dict['over'] = math.floor(line)
    entry_dict['runs'] = 0
    entry_dict['wicket'] = 0
    over = 0
    for line in d:
        
        for k,v in line.items():
        
            if math.floor(k) > over:
                over += 1
                over_dict.append(entry_dict)
                entry_dict = {}
                entry_dict['runs'] = 0
                entry_dict['wicket'] = 0

            entry_dict['over'] = over + 1 
        
            for k2, v2 in v.items():
            #print(v2)
                if k2 == 'bowler':
                    entry_dict['bowler'] = v2
                if k2 == 'runs':
                    entry_dict['runs'] += v2['total']
                if k2 == 'wicket' and v2['kind'] != 'run out':
                    entry_dict['wicket'] += 1 

                #print(over_dict)
    if '2nd innings' in match_list[i]["innings"]:
        d = match_list[i]["innings"][1]['2nd innings']['deliveries']
    else:
        continue
        
    entry_dict_2 = {}
    entry_dict_2['runs'] = 0
    entry_dict_2['wicket'] = 0
    over = 0
    for line in d:
        
        for k,v in line.items():
        
            if math.floor(k) > over:
                over += 1
                over_dict.append(entry_dict_2)
                entry_dict_2 = {}
                entry_dict_2['runs'] = 0
                entry_dict_2['wicket'] = 0

            entry_dict_2['over'] = over + 1 
        
            for k2, v2 in v.items():
            #print(v2)
                if k2 == 'bowler':
                    entry_dict_2['bowler'] = v2
                if k2 == 'runs':
                    entry_dict_2['runs'] += v2['total']
                if k2 == 'wicket' and v2['kind'] != 'run out':
                    entry_dict_2['wicket'] += 1 

                
                
                
df = pd.DataFrame(over_dict)
print(df)
```

```python
df_cp = df
#df_cp.groupby(['bowler'])
df_cp.to_csv('odiout.csv',index= False)
print(df_cp)
```

```python
import powerlaw
data = og_data['runs'].value_counts().sort_index()
results = powerlaw.Fit(data)
print(results.power_law.alpha)
print(results.power_law.xmin)
R, p = results.distribution_compare('power_law', 'lognormal')
print(p,R)
R, p = results.distribution_compare('power_law', 'exponential')
print(p,R)
R, p = results.distribution_compare('exponential', 'lognormal')
print(p,R)    
```
