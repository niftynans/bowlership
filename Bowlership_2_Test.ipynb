```python
bnames = og_data['bowler'].unique()
bdfdict = {}
arr = []
arr2 = []
normbowl = 0
for name in bnames:
    bdfdict[name] = og_data.loc[og_data['bowler'] == name]
    if(len(bdfdict[name])>7):
        if(scipy.stats.normaltest(bdfdict[name]['runs'])[1]>0.001):
            normbowl += 1
            arr2.append(name)
            arr.append(scipy.stats.normaltest(bdfdict[name]['runs'])[0])
print(len(bnames))
print(normbowl)
#plt.plot(arr2, arr, 'o', color='red')
```

```python
#Dâ€™Agostinoâ€™s K^2 Test

notnormcount = 0
maybenormcount = 0
for name in bdfdict:
    
    if(len(bdfdict[name]['runs'])>=8):
        
        print(name)
        print(scipy.stats.normaltest(bdfdict[name]['runs']))
        
        if(scipy.stats.normaltest(bdfdict[name]['runs'])[1] < 0.05):
            print("Data is not normal")
            notnormcount += 1
        else:
            print("Data may be normal")
            maybenormcount += 1
print(notnormcount, maybenormcount)
```

```python
#Shapiro-Wilk Test

notnormcount = 0
maybenormcount = 0

for name in bdfdict:
    
    if(len(bdfdict[name]['runs'])>=8):
        
        print(name)
        print(scipy.stats.shapiro(bdfdict[name]['runs']))
        
        if(scipy.stats.shapiro(bdfdict[name]['runs'])[1] < 0.05):
            print("Data is not normal")
            notnormcount += 1
        else:
            print("Data may be normal")
            maybenormcount += 1
print(notnormcount, maybenormcount)
```

```python
#Anderson-Darling Test
from scipy.stats import anderson

notnormcount = 0
maybenormcount = 0

for name in bdfdict:
    
    if(len(bdfdict[name]['runs'])>=8):
        
        print(name)
        
        data = bdfdict[name]['runs']
        result = anderson(data)
        
        print('Statistic: %.3f' % result.statistic)
        p = 0
        notnormflag = False
        
        for i in range(len(result.critical_values)):
            
            sl, cv = result.significance_level[i], result.critical_values[i]
            
            if result.statistic < result.critical_values[i]:
                print('%.3f: %.3f, data may be normal' % (sl, cv))
                notnormflag = False
                
            else:
                print('%.3f: %.3f, data is not normal' % (sl, cv))
                notnormflag = True
                break
            
        if notnormflag:
            notnormcount += 1
        else:
            maybenormcount += 1
        
print(notnormcount, maybenormcount)        
```
